```python
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

sb.pairplot(df, hue="species", palette="viridis", markers=["o", "s", "D"])
plt.suptitle("Pairwise Scatter Plots of Iris Features", y=1.02)
plt.show()
```

```python
dummies = pd.get_dummies(df.species)

dummies = dummies.astype(int)

dummies.tail()
```

```python
merged_data = pd.concat([df,dummies], axis=1)

merged_data.head()
```

```python
final_data = merged_data.drop(columns='species')

final_data.head()
```

```python
final_data.isnull().sum().sum()
```

```python
t = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
y = final_data[t]
x = final_data.drop(t, axis = 1)
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
print("Train Accuracy of Decision Tree Algorithm ", dt.score(x_train,y_train)*100)
print("Test Accuracy score of Decision Tree Algorithm ", dt.score(x_test,y_test)*100)
```

```python
from sklearn. ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
print("Train Accuracy of Random Forest Algorithm ", rf.score(x_train,y_train)*100)
print("Test Accuracy score of Random Forest Algorithm ", rf.score(x_test,y_test)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
print("Train Accuracy of KNN Algorithm ", knn.score(x_train,y_train)*100)
print("Test Accuracy score of KNN Algorithm ", knn.score(x_test,y_test)*100)
```

```python
# Using Linear Regression model to train on the dataset
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


model = LinearRegression()
model.fit(x_train, y_train)  # Train the model with the training data

# 5. Prediction and Evaluation
# Making predictions on the test set
y_pred = model.predict(x_test)
```

```python
# Calculating Mean Squared Error (MSE) and RÂ² score to evaluate the model performance
mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
r2 = r2_score(y_test, y_pred)  # RÂ² Score (explained variance)

# Printing the evaluation metrics
print(f"Mean Squared Error: {mse}")
print(f"RÂ² Score: {r2}")
```

```python
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("Actual Petal Width")
plt.ylabel("Predicted Petal Width")
plt.title("Actual vs Predicted Petal Width")
plt.show()
```
